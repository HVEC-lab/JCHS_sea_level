##### Data IPCC

```python
path = r'./Data'
name = 'IPCC.db'
name = os.path.join(path, name)
cnxn = sq.connect(name, detect_types = True)
```

```python
names = tuple([nm.upper() for nm in names])
names
```

```python
sql = (
    "SELECT year, name, median*1000 as median FROM data "
    "WHERE process = 'totalrates' "
    "AND scenario = 'ssp245' "
    f"AND name IN {names}"
)
ipcc = pd.read_sql(sql, cnxn)
ipcc
```

```python
# Set capitalization
ipcc['name'] = ipcc['name'].str.title()
ipcc.loc[ipcc['name'] == "Ijmuiden", 'name'] = 'IJmuiden'
```

```python
ipcc
```

```python
fig, ax = plt.subplots(figsize = (20, 8))
ipcc.groupby('name').apply(lambda df: df.plot(x = 'year', y = 'median', ax = ax, label = df['name'].unique().item(), linewidth = 3))
ax.set_xlabel('Year')
ax.set_ylabel('Rate of sea level rise (mm/year)')
fig.tight_layout()
fig.savefig(f'{PICTURES}/ipcc_rates_in_time.jpg');
```
